```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
r_tr, th_tr, ph_tr = np.loadtxt("ray_data.dat", unpack = True)

cond = r_tr == 0.0
```

```python
print(len(cond))
```

```python
from mpl_toolkits.mplot3d import Axes3D

r_dat = []
th_dat = []
ph_dat = []

r_dat2 = []

i = 0
counter = 0

for r in r_tr:
    if r == 0 and counter == 0: break
    r_dat.append(r_tr[i])
    th_dat.append(th_tr[i])
    ph_dat.append(ph_tr[i])
    r_dat2.append(-r_tr[i])
    i = i + 1
    if r == 0: counter = counter + 1
    

print(r_dat)

fig = plt.figure(figsize=(10,10))
ax = fig.gca(projection='3d')

#ax.set_aspect('auto')

#ax.set_zlim(zmin = -fig_lim, zmax = fig_lim)
#ax.set_xlim(xmin = -fig_lim, xmax = fig_lim)
#ax.set_ylim(ymin = -fig_lim, ymax = fig_lim)

#ax.set_zbound(5, fig_lim)
#ax.set_xbound(-fig_lim, fig_lim)
#ax.set_ybound(-fig_lim, fig_lim)
#ax.set_box_aspect((1, 1, 1))

z = list(r_dat * np.cos(th_dat))
y = list(r_dat * np.sin(th_dat) * np.sin(ph_dat))
x = list(r_dat * np.sin(th_dat) * np.cos(ph_dat))

print(x)

# draw sphere
u, v = np.mgrid[0:2*np.pi:20j, 0:np.pi:10j]
r_bh = 1
x_bh = r_bh * np.cos(u)*np.sin(v)
y_bh = r_bh * np.sin(u)*np.sin(v)
z_bh = r_bh * np.cos(v)
ax.plot_wireframe(x_bh, y_bh, z_bh, color="black")

ax.scatter3D(x, y, z, s = 0.1)

ax.view_init(45, 90)

ax.set_xlim(-500,500)
ax.set_ylim(-500,500)
ax.set_zlim(-500,500)


plt.show()
```
